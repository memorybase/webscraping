## Step 2 - CURL

Using https://sqqihao.github.io/trillworks.html as the preferred curl converter

```python
headers = {
    '^sec-ch-ua': '^\\^Not/A)Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^126^\\^, ^\\^Google',
    'Referer': 'https://www.autolist.com/listings',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36',
    '^sec-ch-ua-platform': '^\\^Windows^\\^^',
}

params = (
    ('ads', 'true'),
    ('include_total_price_change', 'true'),
    ('include_time_on_market', 'true'),
    ('include_relative_price_difference', 'true'),
    ('latitude', '34.0549076'),
    ('limit', '20'),
    ('longitude', '-118.242643'),
    ('make', 'Tesla'),
    ('page', '1'),
    ('radius', '100'),
)

response = requests.get('https://www.autolist.com/api/v2/search', headers=headers, params=params)
```

## Step 7 - Put everything together - Loop through results and append data inside a list

Using a for Loop, Loop through all results and append the appropriate data points for all 20 cars 

```python
model = []
mileage = []
year = []
dealer_name = []
price = []

for result in result_items:
    
    # model
    model.append(result['model_name'])
    
    # mileage
    mileage.append(result['mileage'])
    
    # year
    year.append(result['year'])
    
    # dealer_name
    dealer_name.append(result['dealer_name'])
    
    # price
    price.append(result['price'])
```

## Step 9 - Pandas DataFrame - Multiple Pages

5 Pages to get 100 results, 50 Pages to 1000 results

```python
headers = {
    '^sec-ch-ua': '^\\^Not/A)Brand^\\^;v=^\\^8^\\^, ^\\^Chromium^\\^;v=^\\^126^\\^, ^\\^Google',
    'Referer': 'https://www.autolist.com/listings',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36',
    '^sec-ch-ua-platform': '^\\^Windows^\\^^',
}

model = []
mileage = []
year = []
dealer_name = []
price = []

for i in range(1,51):
    
    params = (
        ('ads', 'true'),
        ('include_total_price_change', 'true'),
        ('include_time_on_market', 'true'),
        ('include_relative_price_difference', 'true'),
        ('latitude', '34.0549076'),
        ('limit', '20'),
        ('longitude', '-118.242643'),
        ('make', 'Tesla'),
        ('page', str(i)),
        ('radius', '100'),
    )
    
    # reponse variable for the get request
    response = requests.get('https://www.autolist.com/api/v2/search', headers=headers, params=params)
    
    # json object
    results_json = response.json()
    
    # result item (20 items for each page)
    result_items = results_json['records']
    
    for result in result_items:
    
        # model
        model.append(result['model_name'])

        # mileage
        mileage.append(result['mileage'])

        # year
        year.append(result['year'])

        # dealer_name
        dealer_name.append(result['dealer_name'])

        # price
        price.append(result['price'])
    
```

```python
tesla_multiple_df = pd.DataFrame({'Model': model, 'Mileage': mileage, 'Year': year, 'Dealer_Name': dealer_name, 'Price': price})
```

```python
tesla_multiple_df
```

```python
tesla_multiple_df.to_excel('tesla_multple_pages.xlsx', index=False)
```
